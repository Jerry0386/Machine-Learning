```python
# Import Libraries
import os
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

# Import Dataset
df = pd.read_csv("C:/Users/Jerry/OneDrive/School/Semester 4/Machine Learning/M12/winequality-red.csv")

# Assign Variables
X = df[['alcohol','quality']]

'''
wcss = {} 
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X) 
    wcss[i] = kmeans.inertia_
     
plt.plot(wcss.keys(), wcss.values(), 'gs-')
plt.xlabel("Values of 'k'")
plt.ylabel('WCSS')
plt.show()
'''
# Assign Algorithm
model = KMeans(n_clusters=4)

# Fit Algorithm to Data
model.fit(X)

#Run Algorithm
model_predict = model.predict(X)
centroids = model.cluster_centers_
print(model.cluster_centers_)

# Plot Centroids and Clusters
plt.figure(figsize=(7,5))
plt.scatter(X['alcohol'], X['quality'], c=model_predict, s=50, cmap='rainbow')
plt.scatter(centroids[:,0], centroids[:,1], c='black', s=200, alpha=1);
```
